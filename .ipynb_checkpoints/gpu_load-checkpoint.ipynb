# Read gpu load data

```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
df=pd.read_csv('gpu.log', delimiter=' ')
```

```python
df
```

```python
mem = df[1].str.replace(r'MiB$', '').astype(int)
load = df.iloc[:, 0]
```

```python
load
```

```python
mem
```

```python
from sklearn import preprocessing

column_maxes = mem.max()

mem_max = column_maxes.max()

normalized_mem = ( mem / mem_max ) * 100
```

```python
normalized_mem
```

```python
t = list(range(0, len(mem)))
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax2 = plt.axes()
ax2.plot(t,load,label="load")
ax2.plot(t,normalized_mem,label='memory')
plt.legend(loc='upper right')

#plt.figure(figsize=(10, 15))
#plt.figure(figsize=(2, 8), dpi=400)
plt.show()
```
