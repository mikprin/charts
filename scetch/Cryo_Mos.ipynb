```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import math
```

```python
Mosfet_IV = pd.read_csv('data/Low_TH.vcsv', delimiter=',',header=None)
```

```python
Mosfet_IV.head(3)
```

```python
Vds = Mosfet_IV.iloc[:,0]
Ids_nmos = Mosfet_IV.iloc[:,1]*1e6
Ids_nmos_lvt = Mosfet_IV.iloc[:,3]*1e6
```

## Plot itself

```python
fig=plt.figure(figsize=(4,6), dpi= 100, facecolor='w', edgecolor='k')
ax2 = plt.axes()

plt.title("TSMC 65nm PDK T=1K simulation")
line_temp_30, = ax2.plot(Vds,Ids_nmos,label="nmos_lvt model",linewidth=2.0)
line_temp_55, = ax2.plot(Vds,Ids_nmos_lvt,label="nmos model",linewidth=2.0)
#ax2.legend(['A simple line'])
plt.legend(loc='upper left')
plt.grid()
plt.xlabel('Vgs,V')
plt.ylabel('Ids,uA')
#plt.figure(figsize=(10, 15))
#plt.figure(figsize=(2, 8), dpi=400)
plt.show()
```
