```python
def plot_series_complex(freq_out,ax2,label="label",normilize=False,db=False):
    if (db == True):
        cplx = 10* np.log10(np.sqrt(freq_out.iloc[:,1]*freq_out.iloc[:,1] + freq_out.iloc[:,2]*freq_out.iloc[:,2]))
    else:
        cplx = np.sqrt(freq_out.iloc[:,1]*freq_out.iloc[:,1] + freq_out.iloc[:,2]*freq_out.iloc[:,2])
    if ( normilize == True ):
        ax2.plot(freq_out.iloc[:,0],cplx/max(cplx),label=label)
    else:
        ax2.plot(freq_out.iloc[:,0],cplx,label=label)
```

```python
fig=plt.figure(figsize=(6,6), dpi= 100, facecolor='w', edgecolor='k')

Temp = [0.2,1,70,273,300]
R = 50

ax2 = plt.axes()

for T in Temp:
    V2_R = (4 * T * 1.3e-23 * R * freq)
    V_R = [math.sqrt(V) for V in V2_R]
    dBm = [ 10*math.log( V*1000 , 10)/R for V in V2_R ]
    ax2.plot(freq,V_R , label=f"T={T}K, R = {R}")
    

plt.legend(loc='lower right')
plt.grid()
ax2.set_xscale('log')
ax2.set_yscale('log')
plt.grid(b=True, which='minor', linestyle='--')
plt.xlabel('Bandwidth,Hz')
plt.ylabel('Noise, Vrms')
plt.suptitle("Johnson-Nyquist noise vs T estimation")
plt.show()
```

```python
fig=plt.figure(figsize=(6,6), dpi= 100, facecolor='w', edgecolor='k')

T = 1
Resist = np.geomspace(1, 500e3, num=100)
f = 1e8

ax2 = plt.axes()


boundary_line = [ 1e-6 for R in Resist ]

#for T in Temp:
#    V2_R = (4 * T * 1.3e-23 * R * freq)
#    V_R = [math.sqrt(V) for V in V2_R]
#    dBm = [ 10*math.log( V*1000 , 10)/R for V in V2_R ]
#    ax2.plot(freq,V_R , label=f"{T}, R = {R}")
V2R = 4 * T * 1e-23 * Resist * f
V_V = [math.sqrt(V) for V in V2R]
ax2.plot(Resist,V_V , label=f"T = {T} , frequency = {f:e}")
ax2.plot(Resist,boundary_line, label=f"noise floor of 1e-6")
plt.legend(loc='lower right')
plt.grid()
ax2.set_xscale('log')
ax2.set_yscale('log')
plt.grid(b=True, which='minor', linestyle='--')
plt.xlabel('Intrinsic output resistance,Ohms')
plt.ylabel('Noise, Vrms')
plt.suptitle("Noise vs resistance")
plt.show()
```

## Solve for N bits

```python
fig=plt.figure(figsize=(6,6), dpi= 100, facecolor='w', edgecolor='k')

T = 1
Resist = np.geomspace(1, 500e3, num=100)

N = 16
nbit = [ i for i in range(2,16) ]

Rout = Resist
f = 1e8

ax2 = plt.axes()

boundary_line = [ 1e-6 for R in Resist ]

for bit in nbit:
    V2_R = (4 * T * 1.3e-23 * 2**(bit-2) * Resist * f )
    #V2_R = (4 * T * 1.3e-23 * R * freq)
    V_R = [math.sqrt(V) for V in V2_R]
    #dBm = [ 10*math.log( V*1000 , 10)/R for V in V2_R ]
    ax2.plot(Resist,V_R , label=f"n_bits = {bit}")
#V2R = 4 * T * 1e-23 * Resist * f
#V_V = [math.sqrt(V) for V in V2R]
#ax2.plot(Resist,V_V , label=f"T = {T} , frequency = {f:e}")
ax2.plot(Resist,boundary_line, label=f"noise floor of 1e-6")
plt.legend(loc='lower right')
plt.grid()
ax2.set_xscale('log')
ax2.set_yscale('log')
plt.grid(b=True, which='minor', linestyle='--')
plt.xlabel('Intrinsic output resistance,Ohms')
plt.ylabel('Noise, Vrms')
plt.suptitle("Noise vs bits for kelvin ladder")
plt.show()
```
